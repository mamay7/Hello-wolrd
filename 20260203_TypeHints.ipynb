```python
import math

# Пример 3:
def calculate_square_root():
    """Функция для вычисления квадратного корня с обработкой ошибок"""

    try:
        # Запрос числа у пользователя
        num = float(input("Введите число для вычисления квадратного корня: "))

        # Проверка на отрицательное число
        if num < 0:
            # Для отрицательных чисел используем комплексные числа
            result = math.sqrt(abs(num))
            print(f"Квадратный корень из {num} равен {result}i (комплексное число)")
        else:
            result = math.sqrt(num)
            print(f"Квадратный корень из {num} равен {result}")

        # Дополнительная информация
        print(f"Округление до 2 знаков: {round(result, 2)}")
        print(f"Целая часть: {math.floor(result)}")
        print(f"Округление вверх: {math.ceil(result)}")

    except ValueError:
        print("Ошибка: Введите корректное число!")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Вызов функции
calculate_square_root()
```

```python
import math

def compare_sqrt_methods():
    """Сравнение различных способов вычисления квадратного корня"""

    numbers = [4, 9, 16, 25, 36, 49, 64, 81, 100]

    print("Сравнение методов вычисления квадратного корня:")
    print("-" * 50)
    print(f"{'Число':<10} {'math.sqrt()':<15} {'pow()':<15} {'** 0.5':<15}")
    print("-" * 50)

    for num in numbers:
        method1 = math.sqrt(num)
        method2 = math.pow(num, 0.5)
        method3 = num ** 0.5

        print(f"{num:<10} {method1:<15.4f} {method2:<15.4f} {method3:<15.4f}")

# Вызов функции сравнения
compare_sqrt_methods()
```

```python
import math

class SquareRootCalculator:
    """Класс для вычисления квадратного корня с различными методами"""

    @staticmethod
    def sqrt_builtin(x):
        """Использование math.sqrt()"""
        return math.sqrt(x)

    @staticmethod
    def sqrt_power(x):
        """Использование возведения в степень 0.5"""
        return x ** 0.5

    @staticmethod
    def sqrt_using_log(x):
        """Вычисление через натуральный логарифм (альтернативный метод)"""
        if x > 0:
            return math.exp(math.log(x) / 2)
        else:
            raise ValueError("Нельзя вычислить логарифм от неположительного числа")

    @staticmethod
    def calculate_with_precision(number, decimal_places=4):
        """Вычисление с заданной точностью"""
        result = math.sqrt(number)
        return round(result, decimal_places)

    @staticmethod
    def batch_calculate(numbers):
        """Вычисление квадратных корней для списка чисел"""
        results = {}
        for num in numbers:
            try:
                results[num] = {
                    'value': math.sqrt(num),
                    'rounded': round(math.sqrt(num), 3)
                }
            except ValueError:
                results[num] = {
                    'value': 'Ошибка: отрицательное число',
                    'rounded': None
                }
        return results

# Использование класса
if __name__ == "__main__":
    calculator = SquareRootCalculator()

    # Пример использования
    test_number = 144
    print(f"math.sqrt({test_number}) = {calculator.sqrt_builtin(test_number)}")
    print(f"{test_number} ** 0.5 = {calculator.sqrt_power(test_number)}")
    print(f"С точностью до 2 знаков: {calculator.calculate_with_precision(test_number, 2)}")

    # Пакетное вычисление
    numbers_list = [4, 9, -16, 25, 36.5]
    results = calculator.batch_calculate(numbers_list)

    print("\nРезультаты пакетного вычисления:")
    for num, res in results.items():
        print(f"√{num} = {res['value']} (округлено: {res['rounded']})")
```
